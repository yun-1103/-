輸入立委臉書粉專內容及選舉政見資料

```python
以台中市立委顏寬恒為例，依照月份來統計數值資料的變化
```

```python
import pandas as pd #引入環境內建函式庫pandas並簡寫其為pd
import re
fb1=pd.read_csv('nysu_10902_2019立委_research.csv')#從電腦端資料夾中讀入csv數據庫並命名
fb2=pd.read_csv('nysu_10902_2019_research_right.csv')#讀入後來騰堯補給的資料
fb=pd.merge(fb1,fb2,how="outer")#將fb1和fb2取聯集
politics = pd.read_csv('9th_legislator_promise.csv')
```

```python
fb#叫出讀入的數據來看看
```

```python
#政見 有哪些立委
politics['姓名'].unique()
```

```python
#總共有哪些立委
fb['page_name'].unique()
```

```python
Yan = fb[fb.page_name == "顏寬恒"]#將有關顏寬恒粉專的資料庫從fb數據庫內抓出
Yan.head(2)#
```

```python
Yan[Yan.created_time_taipei == max(Yan.created_time_taipei)]#時間最晚的貼文
```

```python
Yan[Yan.created_time_taipei == min(Yan.created_time_taipei)]#時間最早的貼文
```

```python
Yan['new_date'] = pd.to_datetime(Yan['created_time_taipei']).dt.date#將資料格created time轉為時間格式且最小單位是日期，並給與此筆資料new_date的名稱
Yan['month_year'] = pd.to_datetime(Yan['new_date']).dt.to_period('M')#將資料轉為時間格式且最小單位是月份
Yan['year']=pd.to_datetime(Yan['new_date']).dt.to_period('Y')#將資料轉為時間格式且最小單位是年
Yan.head(1)
```

```python
Yan['post_hour'] = pd.to_datetime(Yan['created_time_taipei']).dt.hour #找出發文 '時間'(hour)
```

```python
#發文時間分佈
post_time = Yan[['post_id','post_hour']].groupby('post_hour').count()
post_time.index.name = 'post_hour'
post_time.reset_index(inplace = True)
post_time
```

```python
results = Yan.groupby('month_year').sum()#將資料按年月分類並加總
results
```

```python
results.index.name = 'newhead' #將上行定義出的results索引標籤命名為newhead
results.reset_index(inplace=True)#將被重新命名過的索引標籤回復成系統本來的排列，個人覺得意義不大
results
```

```python
results['newhead'].map(str)#將索引標籤newhead那欄中所有資料轉為字串格式
```

```python
import matplotlib.pyplot as plt #引入環境內建matplotlib.pyplot繪圖函式庫

labels = results['newhead'].map(str) #將字串化的results['newhead']欄為引入為x軸
width = 0.5       # 柱狀圖中每根柱的寬度

fig, ax = plt.subplots()#在圖中分割出子圖
fig.set_size_inches(15, 8)#整張圖大小為15*8

ax.bar(labels, results['like_count'], width, label='like_count')
ax.bar(labels, results['share_count'], width, label='share_count', bottom=results['like_count'])
ax.bar(labels, results['comment_count'], width, label='comment_count', bottom=results['share_count']+results['like_count'])

ax.legend()

plt.show()
```

```python
September=Yan[Yan.month_year == "2019-09"]#將2019九月的資料抓出並命名為September
```

```python
results_sep = September.groupby('new_date').sum()#將September中的資料按日期加總
results_sep
```

```python
results_sep.index.name = 'newhead'#前幾行有
results_sep.reset_index(inplace=True)
results_sep
```

```python
labels = [str(i) for i in range(1,22)]

fig,ax= plt.subplots()
fig.set_size_inches(15, 8)

ax.plot(labels, results_sep['like_count'])
ax.plot(labels, results_sep['share_count'])
ax.plot(labels, results_sep['comment_count'])

ax.legend()

plt.show()
```

```python
# 每個月按讚、評論、分享總數
results = Yan.groupby('month_year').sum()
results.index.name = 'newhead'
results.reset_index(inplace=True)
results
```

```python
#顏寬恒的最多讚數前50的文章 種類type
toplike_50 = Yan.sort_values(by = ['like_count'],ascending=False).head(50)
toplike_50.groupby('type').count()
```

```python
# 貼文總按讚數最高 
like_top10 = Yan.sort_values('like_count',ascending=False).head(10)
like_top10 = like_top10[['post_id','message','like_count','type']]
like_top10 = like_top10.replace(r'\n',' ',regex=True) # message text preprocessing
like_top10
```

```python
#按讚次數最少的貼文 
lastlike_10 = Yan.sort_values(by = ['like_count'],ascending = True).head(10)
lastlike_10[['message','like_count','month_year','type']].replace(r'\n',' ',regex=True)
```

```python
#從列出評論數（互動數）最高的一天
comment_top20 = Yan.sort_values('comment_count',ascending=False).head(10)
comment_top20 = comment_top20[['message','comment_count','type']]
comment_top20 = comment_top20.replace(r'\n',' ',regex=True) # message text preprocessing
comment_top20
```

```python
#按讚次數最少的貼文 
lesscomment_10 = Yan.sort_values(by = ['comment_count'],ascending = True).head(10)
lesscomment_10[['message','comment_count','month_year']].replace(r'\n',' ',regex=True)
```

```python
# 整年 貼文分享數最高 
share_top10 = Yan.sort_values('share_count',ascending=False).head(10)
share_top10 = share_top10[['message','share_count','type']]
share_top10 = share_top10.replace(r'\n',' ',regex=True) # message text preprocessing
share_top10
```
